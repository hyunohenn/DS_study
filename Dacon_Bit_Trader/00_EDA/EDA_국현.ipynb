## 특징 찾기

```python
coin = 0
df_x = train_x_df[train_x_df.coin_index==coin]
df_y = train_y_df[train_y_df.coin_index==coin]
```

```python
# x, y 데이터의 분산??? 간단 비교
cases = []
for s_i in df_x.sample_id.unique():
    open_vals_x = df_x[df_x.sample_id==s_i].open
    max_val_x = open_vals_x.max()
    min_val_x = open_vals_x.min()
    
    open_vals_y = df_y[df_y.sample_id==s_i].open
    max_val_y = open_vals_y.max()
    min_val_y = open_vals_y.min()
    
    
    case = 1 if (min_val_y > min_val_x and max_val_y < max_val_x) else 0
    cases.append(case)
print(f'{sum(cases)/len(cases)}')
```
