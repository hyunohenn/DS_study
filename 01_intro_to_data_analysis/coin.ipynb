## 7번 코인 1번 샘플만 보자

```python
df = train_x_df[train_x_df.coin_index==7][train_x_df.sample_id==0]
```

```python
df.info()
```

```python
df.head()
```

### 그래프라도 좀 그려볼까

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
plt.plot(df.high, color='skyblue', label='high', alpha=0.5)
plt.plot(df.low, color='darkblue', label='low', alpha=0.5)
plt.plot(df.open, color='darkgreen', label='open', alpha=0.5)
plt.plot(df.close, color='lightgreen', label='close', alpha=0.5)
plt.grid()
plt.legend()
```

```python
i = 1
plt.figure(figsize=(15,10))
for col_name in ['open',  'close', 'high', 'low',]:
    plt.subplot(2, 2, i)
    plt.plot(df.time, df[col_name], color='skyblue')
    plt.grid()
    plt.title(col_name)
    plt.xlabel('time')
    i += 1
```

다 똑같아 보인다!

```python
i = 1
plt.figure(figsize=(20,10))
for col_name in ['volume', 'quote_av', 'trades', 'tb_base_av', 'tb_quote_av',]:
    plt.subplot(3, 2, i)
    plt.plot(df.time, df[col_name], color='skyblue')
    plt.grid()
    plt.title(col_name)
    i += 1
```

```python
# 해당 시간 거래량으로 다음 시간 가격을 예측할 수 있을까?
import seaborn as sns
plt.figure(figsize=(10,5))
sns.regplot(x=df.trades.shift(), y=df.open)
```

```python
df.sort_values('trades', ascending=False).head(10)
```

```python
temp_df = df.sort_values('trades', ascending=False)[5:]
plt.figure(figsize=(10,5))
sns.regplot(x=temp_df.trades.shift(), y=temp_df.open)
```

```python
temp_df = df.sort_values('trades', ascending=False)[5:]
plt.figure(figsize=(10,5))
sns.regplot(x=temp_df.trades.shift(10), y=temp_df.open)
```
